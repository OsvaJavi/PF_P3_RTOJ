```python
import tensorflow as tf
import numpy as np

def entrenar_modelo():
  celcius = np.array([-40, -10, 0, 8, 15, 22, 30], dtype=float)
  fahrenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)

  capa = tf.keras.layers.Dense(units=1, input_shape=[1])
  modelo = tf.keras.Sequential([capa])

  modelo.compile(
      optimizer=tf.keras.optimizers.Adam(0.1),
      loss='mean_squared_error'
  )

  print("Entrenando el modelo...")
  historial = modelo.fit(celcius, fahrenheit, epochs=500, verbose=False)
  print("Modelo entrenado")
  return modelo, historial
```

```python
import modelo

def prediccion(modelo):
  resultado = modelo.predict(100)
  print(f"La temperatura {100.0}°C es igual a {resultado}°f")

  if __name__ == "__main__":
    modelo, historial = modelo.entrenar_modelo()
    prediccion(modelo)
```
